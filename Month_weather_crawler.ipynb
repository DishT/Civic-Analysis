# 1. Get the request from web

```python
import os.path
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
import csv
import numpy as np
from datetime import datetime



def get_web(mm):
    headers = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36', "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8","accept-encoding":"gzip, deflate, br","accept-language":"en-US,zh-TW,zh;q=0.8;q=0.6,en;q=0.4,zh-CN;q=0.2"}
    session = requests.Session()
    re = session.get("https://www.wunderground.com/history/airport/KNYC/2016/{}/29/MonthlyHistory.html?&reqdb.zip=&reqdb.magic=&reqdb.wmo=".format(mm),headers = headers)

    soup = BeautifulSoup(re.text,"html.parser")
    div = soup.find(id = "observations_details")
#text = div.get_text()
#text = text.strip("\n").split("\n")
    return div
a = get_web("01")
```

# 2. Catch the title, subtitle, content

```python
def catch_the_data(div):
    # 2-1. Title in <th>
    title = div.findAll("th",{"class":"nobr"})
    title_list =[]
    for i in title:
        title_list.append(i.get_text()+","+" "+","+" ")
    # 2-2. Subtitle in <td>
    title_sub = div.findAll("td",{"class": {"daily-td-grey","daily-td-white"}})
    title_sub_list = []
    for i in title_sub:
        title_sub_list.append(i.get_text())
    # 2-3. Content in <span>
    content = div.findAll("span",{"class":"wx-value"})
    content_list = []
    for item in content:
        content_list.append(item.get_text())
    print("check: {}/19 = ".format(len(content_list)), len(content_list)/19)
    return (title_list,title_sub_list, content_list)
catch_the_data(a)
```

# 5. Output as .csv

```python
mm = "09"
def write_file(mm,title_list,title_sub_list,content_list):
    f = open('2016{}.csv'.format(mm), 'w')
    # 5.0 initial the data
    count = 0
    text = content_list
    line =""
    
    # 5.1 Write the Title
    
    for i in title_list:
        line = line + i + ","
    line = line + "\n"
    print(line)
    f.write(line)
        
    # Clean the line
    line =""
    # 5.2 Write the sub title
    for i in title_sub_list:
        line = line + i + ","
    line = mm +","+ line + "\n"
    print(line)
    f.write(line)

    line =""
    # Counter for days 
    day = 1
    for i in range (len(text)): 
        if count != 18:
            line = line + text[i] +","
            count = count + 1
            #print(i,count)
        
        else:
            line = "2016/{}/{}".format(mm,day)+ "," + line + text[i]+"\n"
            print(line+"\n")
            f.write(line)
            line = ""
            count = 0
            day += 1
    f.close()
```

# Main Program

```python

for mm in range(1,13):
    if mm < 10:
        mm = "0" + str(mm)
    else:
        mm = str(mm)
    # 1. Get the request from web
    web_html = get_web(mm)
    content = catch_the_data(web_html)
    #print(content)
    write_file(mm,content[0],content[1],content[2])
    
```
